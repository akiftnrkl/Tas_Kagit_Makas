```python
######### OYUN TANITIMI ######### 

# Gerekli kütüphaneleri ekliyoruz
import random
import time
import emoji

# Emoji karakterlerini tanıtıyoruz
kikirdayan_yuz = "😂"
uzgun_yuz = "😞"
kizgin_gulumseme = "😡"
saka_yapma_yuz = "😜"

# Oyunun tanıtımını yapıyoruz
print("\nMerhaba! Taş-Kağıt-Makas oyununa hoş geldin!")
time.sleep(1)
print("\nOyunun Kuralları:")
time.sleep(1)
print("\n-Taş, kağıt, makastan birini seçiceksin. Duruma göre ikimizden birisi kazanacak.")
time.sleep(1)
print("-Çoğunlukla ben kazanacağım :)")
time.sleep(1)
print("\n-İşte hangi durumda kimin kazanacağına dair bir liste... ")
time.sleep(1)
print("\nTaş vs. Kağıt: Kağıt kazanır")
print("Taş vs. Makas: Taş kazanır")
print("Taş vs. Taş: Beraberlik")
print("\nKağıt vs. Taş: Kağıt kazanır")
print("Kağıt vs. Makas: Makas kazanır")
print("Kağıt vs. Kağıt: Beraberlik")
print("\nMakas vs. Taş: Taş kazanır")
print("Makas vs. Kağıt: Makas kazanır")
print("Makas vs. Makas: Beraberlik")
print("\nAnladın değil mi? Sanmıyorum ama neyse.")
time.sleep(1)
print("\nBu arada 2 tur kazanan o oyunu kazanır.")

######### OYNAMAK İSTER MİSİN? #########  

# Kullanıcıya oynamak isteyip istememesi soruluyor
print("\nOynamak istiyor musun? (evet/hayır)")
cevap = input().lower()

# Kullanıcının verdiği cevaba göre bilgisayarın yanıtını yazıyoruz
while True:
    if cevap == "evet":
        print("\nSen kaşındın...")
        break
    elif cevap == "hayır":
        print("\nO zaman beni neden çağırdın??")
        break
    else:
        print("\nKör müsün? evet veya hayır yaz!!", kizgin_gulumseme)
        cevap = input().lower()

######### OYUN DÖNGÜSÜNÜ BAŞLATIYORUZ #########  

if cevap == "evet":
    def tas_kağıt_makas_mehmet_akif_tanrikulu():
        # İhtimalleri yazıyoruz
        ihtimaller = ['taş', 'kağıt', 'makas']

        # Oyun sayacını başlatıyoruz
        oynanacak_oyun_sayisi = 1

        while True:
            oyuncu_skor = 0
            bilgisayar_skor = 0
            oynanan_tur_sayisi = 1

            while oyuncu_skor < 2 and bilgisayar_skor < 2:
                # Bilgisayarın seçim yapmasını sağlıyoruz
                bilgisayar_secim = random.choice(ihtimaller)

                # Oyunu belirtiyoruz
                print(f"{oynanacak_oyun_sayisi}. OYUN")

                # Turu belirtiyoruz
                print(f"######  {oynanan_tur_sayisi}. TUR ######")

                # Kullanıcının seçim yapmasını sağlıyoruz
                oyuncu_secim = input("Hangisini seçiyorsun? (taş, kağıt, makas)\n").lower()

                # Kullanıcının geçerli bir seçim yapıp yapmadığını kontrol ediyoruz
                if oyuncu_secim not in ihtimaller:
                    print("\nNe diyorsun be! Okuma yazman yok mu?", kizgin_gulumseme)
                    continue

                # Kazanma durumları sözlüğü
                kazananlar = {
                    ('taş', 'kağıt'): 2,
                    ('taş', 'makas'): 1,
                    ('kağıt', 'taş'): 1,
                    ('kağıt', 'makas'): 2,
                    ('makas', 'taş'): 2,
                    ('makas', 'kağıt'): 1
                }

                if oyuncu_secim == bilgisayar_secim:
                    print(f"\n-Senin seçimin = {oyuncu_secim}")
                    print(f"-Bilgisayarın seçimi = {bilgisayar_secim}")
                    print("-Tur Berabere!\n")
                else:
                    sonuc = kazananlar.get((oyuncu_secim, bilgisayar_secim))
                    if sonuc == 1:
                        print(f"\n-Senin seçimin = {oyuncu_secim}")
                        print(f"-Bilgisayarın Seçimi = {bilgisayar_secim}")
                        print("-Turu sen kazandın!\n")
                        oyuncu_skor += 1
                    elif sonuc == 2:
                        print(f"\n-Senin seçimin = {oyuncu_secim}")
                        print(f"-Bilgisayarın seçimi = {bilgisayar_secim}")
                        print("-Turu bilgisayar kazandı!\n")
                        bilgisayar_skor += 1

                oynanan_tur_sayisi += 1
                print(f"Bilgisayar: {bilgisayar_skor} | Sen: {oyuncu_skor}\n")

            # Oyun bittiğinde son durumu yazdırıyoruz
            if oyuncu_skor == 2:
                print("\nSen kazandın", uzgun_yuz)
            elif bilgisayar_skor == 2:
                print("\nTabii ki ben kazandım!", kikirdayan_yuz)

            oynanacak_oyun_sayisi += 1

            while True:
                # Bilgisayarın bir sonraki oyun için karar vermesi
                bilgisayar_tercih_list = ['yes', 'no']
                bilgisayar_tercih = random.choice(bilgisayar_tercih_list)

                # Yeni bir oyuna geçmek için kullanıcıdan onay alıyoruz
                cevap = input("\nBir oyun daha oynamak ister misin? (evet/hayır)\n").lower()

                # Kullanıcının verdiği cevaba göre ilerliyoruz
                if cevap == "evet" and bilgisayar_tercih == "yes":
                    print("\nOynayalım o zaman! Sonra ağlamazsın umarım. Seninle uğraşamam çünkü :)", saka_yapma_yuz)
                    break  # Yeni oyuna geçmek için döngüyü kırıyoruz
                elif cevap == "evet" and bilgisayar_tercih == "no":
                    print("\nÇok işim var. Ben oynamak istemiyorum.")
                    return  # Oyun döngüsünden çıkıyoruz
                elif cevap == "hayır":
                    print("\nTamam, sonra görüşürüz!")
                    return  # Oyun döngüsünden çıkıyoruz
                else:
                    print("\nNe diyorsun be! Evet veya hayır yazmak bu kadar zor mu?", kizgin_gulumseme)

    # Fonksiyonu çağırarak oyunu başlatıyoruz
    tas_kağıt_makas_mehmet_akif_tanrikulu()
```
